```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, MultiHeadAttention, LayerNormalization, Dropout, Flatten
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("real_time_S&P_500.csv")
print(df.rows)  # Check column names
```

```python
df = pd.read_csv("real_time_S&P_500.csv", parse_dates=["Date"], index_col="Date")
```

```python
# === Feature Engineering ===
sequence_length = 90
scaler = MinMaxScaler()
df['Scaled_Volatility'] = scaler.fit_transform(df[['Log_Returns']])

def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

X, y = create_sequences(df['Scaled_Volatility'].values, sequence_length)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
# === Build Transformer Model ===
def transformer_block(inputs, num_heads=4, ff_dim=32):
    attn_output = MultiHeadAttention(num_heads=num_heads, key_dim=ff_dim)(inputs, inputs)
    attn_output = Dropout(0.1)(attn_output)
    out1 = LayerNormalization(epsilon=1e-6)(inputs + attn_output)
    ffn_output = Dense(ff_dim, activation='relu')(out1)
    ffn_output = Dense(inputs.shape[-1])(ffn_output)
    ffn_output = Dropout(0.1)(ffn_output)
    return LayerNormalization(epsilon=1e-6)(out1 + ffn_output)

inputs = Input(shape=(sequence_length, 1))
x = transformer_block(inputs)
x = Flatten()(x)
x = Dense(64, activation='relu')(x)
x = Dense(1)(x)
model = Model(inputs=inputs, outputs=x)
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python

# === Train Model ===
history = model.fit(X_train, y_train, epochs=100, batch_size=16, validation_data=(X_test, y_test), verbose=1)
```

```python
# === Make Predictions ===
predicted_volatility = model.predict(X_test)
predicted_volatility = scaler.inverse_transform(predicted_volatility.reshape(-1, 1))
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python

# Save Predictions
df_transformer_results = pd.DataFrame({
    "Date": df.index[-len(y_test):],
    "Predicted Volatility (Transformer)": predicted_volatility.flatten(),
    "Actual Volatility": y_test_actual.flatten()
})
df_transformer_results.set_index("Date", inplace=True)
df_transformer_results.to_csv("transformer_volatility_predictions.csv")
```

```python

# Plot Results
plt.figure(figsize=(12, 6))
plt.plot(df_transformer_results.index, df_transformer_results["Actual Volatility"], label="Actual Volatility", color="red")
plt.plot(df_transformer_results.index, df_transformer_results["Predicted Volatility (Transformer)"], label="Predicted Volatility (Transformer)", linestyle="dashed", color="blue")
plt.title("Transformer Model Forecast vs. Actual Volatility")
plt.legend()
plt.show()

print("✅ Transformer-based volatility model trained and predictions saved!")
```
